```python
import pandas as pd
import numpy as np
import datetime
import seaborn as sns
sns.set_theme(style="darkgrid")
sns.set({"figure.figsize":(22,15)})
sns.set(font_scale = 1.15)
import plotly.express as px
import seaborn as sns
import pandas as pd
import plotly.graph_objects as go
import matplotlib.pyplot as plt
```

```python
f1 = pd.read_csv("../../Data/pilotosf1.csv",encoding = "ISO-8859-1")
bk = pd.read_csv("../../Data/baloncesto.csv",encoding = "ISO-8859-1")
tn = pd.read_csv("../../Data/tenistas.csv",encoding = "ISO-8859-1")
vl = pd.read_csv("../../Data/voley.csv",encoding = "ISO-8859-1")
```

```python
df = pd.concat([f1, bk, tn, vl], axis=0)
```

```python
x,y = 'Deporte', 'trimestre'

(df
.groupby(x)[y]
.value_counts(normalize=True)
.mul(100)
.rename('percent')
.reset_index()
.pipe((sns.catplot,'data'), x=x,y='percent',hue=y,kind='bar',height=12))
```

```python
df.to_csv('../../Data/otrosdeportes.csv', encoding='utf-8')
```
